```python
# import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a pandas series object
x = pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
company_name =  ['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert']
```

```python
# plot a pie chart to show the share of each company
plt.title('Name of the company with respect to X')
plt.pie(x,labels = company_name,autopct='%.1f%%')
plt.figure(figsize=(7,10))
plt.show();
```

```python
# Draw a box plot to visualize the outliers
sns.boxplot(x);
plt.show()
```

```python
# finding out the outliers and calculate  μ,σ,σ^2
```

```python
Q1 = np.quantile(x,0.25)
Q3 = np.quantile(x,0.75)
med = np.median(x)
IQR = Q3 - Q1
upper_bound = Q3+(1.5*IQR)
lower_bound = Q1-(1.5*IQR)
```

```python
Outliers = x[(x <= lower_bound) | (x >= upper_bound)]
print('The outlier in the boxplot:',Outliers)
```

```python
print("Mean=",x.mean())
print("Standars deviation=",x.std())
print("Variance=",x.var())
```
